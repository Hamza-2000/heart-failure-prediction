```python
import pandas as pd
import statsmodels.api as sm
import pylab as py
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
dataset=pd.read_csv("C:\\Users\\HAMZA\\Desktop\\heart_failure_clinical_records_dataset.csv")
```

```python
deaths=dataset[dataset.DEATH_EVENT == 1]
deaths.ejection_fraction.hist()
plt.title('histogramme de la variable ejection_fraction en fonction de nombres de morts')
plt.xlabel('Le poucentage de sang pompée chaque fois que le coeur contracte (+55 est normale)')
plt.ylabel('Nombre de morts')
```

```python
sm.qqplot(dataset.platelets,line='q')
py.show()
```

```python
correlation=dataset.corr()
plt.figure(figsize=(15, 10))
sns.heatmap(correlation, annot=True)
```

```python
X = dataset.loc[:, dataset.columns != 'DEATH_EVENT']
y = dataset.loc[:, dataset.columns == 'DEATH_EVENT']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=5)
```

```python
lr = LogisticRegression()
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
print('La précision du modèle de regression logistique : {:.2f}'.format(lr.score(X_test, y_test)))
```
